```python
#calculate % of null values in each columns
total = inp0.isnull().sum().sort_values(ascending = False)
percent = (inp0.isnull().sum()/inp0.isnull().count()*100).sort_values(ascending = False)
missing_inp0  = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
missing_inp0.head(50)
```

```python
#list of all columns whose % is more than 35
nullCol = list(missing_inp0[missing_inp0.Percent>=40.00].index)
nullCol
```

```python
#dropping columns whose null values > 40.00% and store it in another dataframe inp1
inp1=inp0.drop(labels=nullCol,axis=1)
```
